```python
import torch
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread("data/images/1RYdYTP1uZg_2000.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(im)
plt.show()
```

```python
im_tensor = torch.from_numpy(im)
im_tensor.shape
```

```python
histogram = torch.histc(im_tensor.float(), bins=256)
histogram.shape
```

```python
plt.bar(range(256), histogram, align='center')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
hist = torch.zeros([256])
var_b = torch.zeros([256])

for i in range(256):
    hist[i] = torch.sum(im_tensor == i)

for i in range(256):
    wb = torch.sum(hist[:i+1])
    wf = torch.sum(hist[i+1:])

    if wb == 0 or wf == 0:
        var_b[i] = -1
    else:
        ub = torch.sum(hist[:i+1] * torch.arange(start=0, end=i+1)) / wb
        uf = torch.sum(hist[i+1:] * torch.arange(start=i+1, end=256)) / wf
        var_b[i] = wb * wf * torch.pow(ub - uf, 2)

out = int(torch.argmax(var_b))
```

```python
out_tensor = torch.zeros(im_tensor.shape)
out_tensor[(im_tensor.float()>out)] = 255
out_tensor = out_tensor.to(torch.uint8)
print(out_tensor.shape)
print(im_tensor.shape)
```

```python
out_np = out_tensor.numpy()
out_np.shape
```

```python
plt.imshow(out_np)
plt.show()
```

```python
sift = cv2.SIFT_create()
kp = sift.detect(out_np, None)
out_np = cv2.drawKeypoints(out_np,kp,out_np)
plt.imshow(out_np)
plt.show()
```

```python
cv2.imwrite('image.png', out_np)
```

```python
img = cv2.imread("data/images/1RYdYTP1uZg_2000.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)
dst = cv2.cornerHarris(gray,2,3,0.04)
 
#result is dilated for marking the corners, not important
dst = cv2.dilate(dst,None)
 
# Threshold for an optimal value, it may vary depending on the image.
img[dst>0.001*dst.max()]=[0,0,255]

plt.imshow(img)
plt.show()
cv2.imwrite('image.png', img)
```
